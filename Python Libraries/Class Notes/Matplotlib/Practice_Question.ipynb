```python
# Task :
# COMPARE TEMPERATURE OF CITIES OVER A WEEK USING SUBPLOTS AND LOOPS

days = ["Mon", "Tue", "Wed", "Thu", "Fri"]

cities = ["New York", "London", "Delhi", "Tokyo"]

temperatures = [
[22, 23, 21, 24, 25],   # New York
[18, 19, 17, 20, 21],   # London
[30, 32, 31, 33, 34],   # Delhi
[25, 26, 24, 27, 28]    # Tokyo

]
```

```python
# MANUAL APPROCH

import matplotlib.pyplot as plt

fig ,ax = plt.subplots(2, 2, figsize = (7,7))

ax[0][0].plot(days,temperatures[0])
ax[0][0].set_title(cities[0])
ax[0][0].set_xlabel("Days")
ax[0][0].set_ylabel("Temperature") 

ax[0][1].plot(days,temperatures[1])
ax[0][1].set_title(cities[1])
ax[0][1].set_xlabel("Days")
ax[0][1].set_ylabel("Temperature") 

ax[1][1].plot(days,temperatures[2])
ax[1][1].set_title(cities[2])
ax[1][1].set_xlabel("Days")
ax[1][1].set_ylabel("Temperature") 

ax[1][0].plot(days,temperatures[3])
ax[1][0].set_title(cities[3])
ax[1][0].set_xlabel("Days")
ax[1][0].set_ylabel("Temperature") 

fig.suptitle("TEMPERATURE OF CITIES OVER A WEEK")
fig.tight_layout()
```

```python
# NOW USINg nested LOOPS
fig ,ax = plt.subplots(2, 2, figsize = (7,7))
x = 0
for i in range(2):
    for j in range(2):
        ax[i][j].plot(days,temperatures[x], marker = "o")
        ax[i][j].set_title(cities[x])
        # ax[i][j].set_xlabel("Days")
        # ax[i][j].set_ylabel("Temperature")
        ax[i][j].grid(True)
        x += 1

fig.supxlabel("Days") # gives a common value to entier fig
fig.supylabel("Temperture")
fig.suptitle("TEMPERATURE OF CITIES OVER A WEEK")
fig.tight_layout()
```

```python
# USING A SINGLE LOOP
import numpy as np

fig ,ax = plt.subplots(2, 2, figsize = (7,7)) # rigth now ax is a 2d array
# converting 2d ax to 1d axes
axes = ax.flatten()

for i, x in enumerate(axes): # enumerate in array gives back index , and value at index
    x.plot(days, temperatures[i], marker = "o")
    x.set_title(cities[i])
    x.grid(True)

fig.supxlabel("Days") 
fig.supylabel("Temperture")
fig.suptitle("TEMPERATURE OF CITIES OVER A WEEK")
fig.tight_layout()
```
