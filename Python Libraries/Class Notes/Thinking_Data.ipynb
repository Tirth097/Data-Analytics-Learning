```python
import json 
```

```python
#load data
def load_data(filename):
    with open (filename,"r") as f :
        data = json.load(f)
    return data
```

```python
data = load_data("Store_Data.json")
print(data,type(data))
```

```python
#clean and structure data 

def clean_data(data):
    cleaned_data = []
    unique_users = set()
    
    #clean ratings - data consistency
    text_to_num = {"one":1,"two":2,"three":3,"four":4,"five":5}
    for users in data :
        raw_rating = str(users["rating"]).strip().lower()
        if raw_rating in text_to_num:
            raw_rating = text_to_num[raw_rating]
        users["rating"] = raw_rating

    # handling missing data
        raw_age = users.get("age")
        if (raw_age == None):
            users["age"] = None 

    # deduplication
        if (users["name"] in unique_users):
            continue
        unique_users.add(users["name"])
        cleaned_data.append(users)
    return cleaned_data
    # print(cleaned_data)
```

```python
data = clean_data(data)
```

```python
# geting meaningful insigts
def insights(data):

    # avg rating
    total_rating = 0
    num = 0
    for users in data :
        total_rating += float(users["rating"])
    print(f"avg_rating = {total_rating/len(data)}")

    #percentage of users with >3 rating
    for users in data:
        if(float(users["rating"])<3):
            num += 1
    print(f"percentage of users no satisfed : {(num/len(users))*100} %")
        
```

```python
insights(data)
```

```python
# Recomendation of brand as per rating
def get_recomendation(data):
    recomendation = []
    for users in data:
        current_recom = {}
        current_recom["name"] = users["name"]
        if (float(users["rating"])>4):
            current_recom["brand"] = "apple"
        else:
            current_recom["brand"] = "samsung"
        recomendation.append(current_recom)
    print (recomendation)
            
```

```python
get_recomendation(data)
```
