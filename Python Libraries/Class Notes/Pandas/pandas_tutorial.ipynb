```python
## FEATURE ENGINEERING ( TRANSFORMING DATA)

# 1} funtions in it

# a} apply function
df2 = df.copy()
df2["tax"] = df2["income"].apply(lambda x : "20 %" if x >= 50000 else "10 %")
df2

# b} gender map
gender_map = {"Female" : "f" , "Male" : "m" , "Unknown" : "u"}
df2["gender"] = df2["gender"].map(gender_map)
df2

# c} assign 
df2 = df2.assign(new_income = df2["income"] * 1.1) # (1 + 0.10) × Income = 1.10 × Income

# d} replace
df2["country"] = df2["country"].replace("USA","US") #("old","new")
df2

# e} rename
df2.columns = ["Id","Name","Age","Country","Gender","Income","Tax","New_Income"]
df2.rename(columns = {"Income":"Salary"})
df2.rename(index = {1:"First"})

#f} sorting values
df2["Income"].sort_values() # get only sorted values
# df2.sort_values("Income") # get all the values as per soarted ones 
# df2.sort_values(["Income","Age"]) # get all the values as per soarted ones but when income same then age check 
df_sorted = df2.sort_values("Income", ascending = False) # if needed desending values

# sort index
df2.sort_index()  

# g} reset index
#df2.reset_index() # it preserves by making to new coloumn
df_sorted.reset_index(drop = True) # it makes changes and new index are assined not in new coloumn


# h} ranking 
df_sorted["Ranks"] = df_sorted["Income"].rank()  # 1st rank to lowest salary [if 5 and 6 rank same then they get 5.5]
df_sorted["Ranks"] = df_sorted["Income"].rank(ascending = False) # 1st rank to highest salary 
df_sorted["Ranks"] = df_sorted["Income"].rank(ascending = False, method = "min") # assign lower rank
df_sorted["Ranks"] = df_sorted["Income"].rank(ascending = False, method = "max") # assign higeher rank  
df_sorted["Ranks"] = df_sorted["Income"].rank(ascending = False, method = "dense")  
df_sorted

# I} reorder
df_sorted[["Id","Name","Age","Country","Gender","Income","New_Income","Tax"]] # just enter in the order that you want
```

```python
# GROUPING AND AGGREGATION OF DATA

df.groupby("country")["income"].mean()
df.groupby("country")["income"].min()
df.groupby("country")["income"].max()
df.groupby("country")["income"].count()

df.groupby("gender")["income"].mean()

#  using multiple aggregate fnx at same time
df.groupby("country")["income"].agg(["mean","max","min","count"])

# assigning coloumn names
df.groupby("country")["income"].agg(avg_salary = "mean",max_salary = "max", number = "count")

# using different agg fnx on different column
df.groupby("country").agg({
    "income" : "mean",
    "age" : "min"
})

# now to give name in this
df.groupby("country").agg(
   avg_salary = ("income" , "mean"),
    min_age = ("age" , "min" )
)

df.groupby("gender").agg(
    avg_salary = ("income","mean"),
    higest_salary = ("income","max"),
    lowset_salary = ("income","min"),
    number = ("id","count")
)
```

```python
# melt and piviot

# melt
df = pd.DataFrame({
    "country":["a","b","c","d"],
    "year": ["1","2","3","4"],
    "sales":["100","200","300","400"],
    "profit":["10","20","30","40"],
})
melted_df = df.melt(
    id_vars = ["country","year"], # coloums to be kept unchanged
    value_vars = ["sales","profit"], # coloums to merge
    var_name = "metrics", # mered coloumn new name 
    value_name = "value" # its values coloumn name
)
```

```python
# pivot

melted_df.pivot(
    index = ["country","year"],
    columns = "metrics",
    values = "value"
)
```

```python
# MERING AND JOINING

customer_df = pd.DataFrame({
    "customer_id" : [1,2,3,4],
    "name" : ["a","b","c","d"]
    
})

order_df = pd.DataFrame({
    "order_id": [101, 102, 103, 104],
    "customer_id": [2, 1, 4, 5],
    "amount": [250, 120, 300, 180]
})

#1} MERGING
#A)inner join
pd.merge(customer_df, order_df, on = "customer_id")

#b)left join
pd.merge(customer_df, order_df, on = "customer_id", how = "left")

#b)right join
pd.merge(customer_df, order_df, on = "customer_id", how = "right")

#b)outer join
pd.merge(customer_df, order_df, on = "customer_id", how = "outer")
```
